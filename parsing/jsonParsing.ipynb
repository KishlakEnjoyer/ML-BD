```python
# –ó–ê–≥—Ä—É–∂–∞–µ–º json —Ñ–∞–π–ª—ã 
all_jsons = glob.glob(r"jsons\*.json")
```

```python
# –°–æ–∑–¥–∞–µ–º —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞
dct = {'name_company':[], 'rate_company':[], 'subscribers':[], 'industries':[], 'about_company':[], 'text':[], 'publish_date':[], 'rate_post':[], 'views_post':[], 'refs_quant':[]}
```

# 3. –ü–∞—Ä—Å–∏–Ω–≥ üì§

```python
import json
import numpy as np
import dateparser

# –ü—Ä–æ—Ö–æ–¥–∏–º—Å—è –ø–æ –≤—Å–µ–º JSON —Ñ–∞–π–ª–∞–º
for json_file in all_jsons:
    with open(json_file, 'r', encoding='utf-8') as file:
        data = json.load(file)

    # –ù–∞–∑–≤–∞–Ω–∏–µ –∫–æ–º–ø–∞–Ω–∏–∏
    name_company = json_file.split('\\')[-1].split('.json')[0]
    
    refs_quant = data.get('refs_quant', 0)
    
    # –ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –∫–æ–º–ø–∞–Ω–∏–∏
    info = data.get('info') or {}
    company_rate = info.get('rate', 0)
    subscribers = info.get('subs', 0)
    industries = ', '.join(info['industries']) if info.get('industries') else "–ù–µ—Ç –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏"
    about_company = info.get('about', "–ù–µ–∏–∑–≤–µ—Å—Ç–Ω–æ")

    for article in data.get('refs', []):
        if not isinstance(article, list) or len(article) < 3:
            continue
            
        text = article[0] if len(article) > 0 else ""
        datetime_dict = article[1] if len(article) > 1 else {}
        article_info = article[2] if len(article) > 2 else {}

        # –î–∞—Ç–∞
        datetime_str = (datetime_dict.get('day', '') + datetime_dict.get('month', '') + datetime_dict.get('time', ''))
        publish_date = dateparser.parse(datetime_str, languages=['ru'])

        ratepost = article_info.get('rate', 0)
        viewspost = article_info.get('views', 0)

        # –ó–∞–ø–æ–ª–Ω–µ–Ω–∏–µ —Å–ª–æ–≤–∞—Ä—è
        dct['name_company'].append(name_company)
        dct['rate_company'].append(company_rate)
        dct['refs_quant'].append(refs_quant)
        dct['subscribers'].append(subscribers)
        dct['industries'].append(industries)
        dct['about_company'].append(about_company)
        dct['text'].append(text)
        dct['publish_date'].append(publish_date)
        dct['rate_post'].append(ratepost)
        dct['views_post'].append(viewspost)
```

```python
df = pd.DataFrame(dct)
df
```
